```python
import pandas as pd

data_URL1 = 'https://storage.googleapis.com/scsu-data-science/movies_imdb.csv'
df = pd.read_csv(data_URL1)
df_1a = df[['director_name', 'movie_title', 'duration']]
df_1a.head(20)
```

Show all the movies directed by a director specified by the user (title, year, and gross sales).

```python
userDirector = input('Enter name of a director: ')

df_1b = df[df['director_name'] == userDirector][['movie_title', 'title_year', 'gross']]
df_1b.head(20)
```

```python
totalGross = df_1b['gross'].sum()
print('The total gross sales for {} is ${:,.2f}'.format(userDirector, totalGross))
```

```python
data_URL2 = 'https://storage.googleapis.com/scsu-data-science/mpg.csv'
df_part2 = pd.read_csv(data_URL2)
df_part2.head()
```

Create a new column make that only contains the make of the car.

```python
df_part2['make'] = df_part2['name'].str.split().str.get(0)
display(df_part2['make'])
```

```python
ave_mpg = df_part2['mpg'].groupby(df_part2['make'])
print(ave_mpg.mean())
```
