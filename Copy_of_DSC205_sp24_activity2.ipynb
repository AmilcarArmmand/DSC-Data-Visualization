## Exercise 1

```python
import pandas as pd

movies = pd.read_csv('https://storage.googleapis.com/scsu-data-science/movies_imdb.csv')
movies.head()
```

```python
# Q1
movies.loc[:19, ['director_name', 'duration', 'movie_title']]
```

```python
list(movies.columns)
```

```python
director = input('Input a director name: ')

# Case-INsensitive match
# Save the director's movies in a new df (df_by_director)
df_by_director = movies.loc[movies['director_name'].str.lower() == director.lower(), ['movie_title', 'title_year', 'gross']]

# Uncomment next line (and comment previous line) if case-sensitive match is preferred
# df_by_director = movies.loc[movies['director_name'] == director, ['movie_title', 'title_year', 'gross']]

df_by_director
```

```python
# Sum the gross column for the director's df
tot_gross = df_by_director['gross'].sum()
print(f"{director}'s films grossed ${tot_gross}.")
```

```python
# Create groups by genre and compute mean duration for each group
movies_grouped = movies['duration'].groupby(movies['genre'])
movies_grouped.mean()
```

## Exercise 2

```python
import pandas as pd

cars = pd.read_csv('https://storage.googleapis.com/scsu-data-science/mpg.csv')
cars.head()
```

```python
# Extract the first word (token) that appears in the 'name' column and put
# it in a new column 'make'
cars['make'] = cars['name'].str.split().str.get(0)
cars.head()
```

```python
# Create groups by car make, and compute average mpg for each group
cars_grouped = cars['mpg'].groupby(cars['make'])
cars_grouped.mean()
```

```python
import pandas as pd

deaths = pd.read_csv('https://storage.googleapis.com/scsu-data-science/CT_accidental_drug_related_deaths.csv')
deaths.head()
```

```python
deaths['Age'].hist(bins=20)
```

```python
deaths.describe()
```

```python
# Count deaths by residence city
deaths_by_city = deaths['Residence City'].value_counts()
deaths_by_city
```

```python
# Output the first entry returned by value_counts.  index contains city names and
# the value corresponding to each index is the deaths count for the city.
print(f"{deaths_by_city.index[0]} is the city with the most deaths ({deaths_by_city[0]}).")
```
