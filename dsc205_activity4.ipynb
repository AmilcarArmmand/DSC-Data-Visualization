```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Goal 1

```python
income_URL = 'https://raw.githubusercontent.com/iantonios/dsc205/main/CT-towns-income-census-2020.csv'
df = pd.read_csv(income_URL, thousands=',')
df.head()
```

```python
# Rename columns to remove spaces and clean up the category name.
df.rename(columns={'Planning\nRegion': 'Region',
                         'Per capita\nincome': 'Per_Capita_Income',
                         'Median\nhousehold\nincome':'Median_Household_Income',
                         'Median\nfamily\nincome': 'Median_Family_Income',
                         'Number of\nhouseholds': 'Households'},
                inplace=True)
df.columns
```

```python
df.dtypes
```

```python
# Remove $ and , from columns with income.
df['Per_Capita_Income'] = df['Per_Capita_Income'].str.replace('$', '').str.replace(',', '')
df['Median_Household_Income'] = df['Median_Household_Income'].str.replace('$','').str.replace(',','')
df['Median_Family_Income'] = df['Median_Family_Income'].str.replace('$','').str.replace(',','')
df.head()
```

```python
df['Per_Capita_Income'] = df['Per_Capita_Income'].astype('int64')
df['Median_Household_Income'] = df['Median_Household_Income'].astype('int64')
df['Median_Family_Income'] = df['Median_Family_Income'].astype('int64')
df.dtypes
```

```python
sns.set(font_scale=0.8)
g1a = sns.catplot(data=df, x='Median_Household_Income', y='County', kind='swarm', s=15)
g1a.set(title='Median Income Distribution for CT Towns by County')
```

```python
sns.set(font_scale=0.8)
g1b = sns.catplot(data=df, x='Median_Family_Income', y='County', kind='swarm', s=15)
g1b.set(title='Median Family Income Distribution for CT Towns by County')
```

```python
sns.set(font_scale=0.8)
g1c = sns.catplot(data=df, x='Per_Capita_Income', y='County', kind='swarm', s=15)
g1c.set(title='Per Capita Income Distribution for CT Towns by County')
```
